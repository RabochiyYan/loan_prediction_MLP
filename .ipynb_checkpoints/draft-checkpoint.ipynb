```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

# Создадим пример данных с выбросами
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)
data = np.append(data, [5, -5, 6, -6])

# Преобразуем данные в DataFrame
df = pd.DataFrame(data, columns=['values'])
df
```

```python

# Используем IsolationForest для обнаружения выбросов
isolation_forest = IsolationForest(contamination=0.04)
df['isolation_forest_outlier'] = isolation_forest.fit_predict(df[['values']])

# Используем LocalOutlierFactor для обнаружения выбросов
lof = LocalOutlierFactor(contamination=0.04)
lof.fit(df[['values']])
df['lof_outlier'] = lof._predict(df[['values']])


df
```

```python
# Удаляем выбросы, определенные IsolationForest
df_cleaned_isolation = df[df['isolation_forest_outlier'] != -1]

# Удаляем выбросы, определенные LOF
df_cleaned_lof = df[df['lof_outlier'] != -1]

# Выводим очищенные данные
print("Данные после удаления выбросов с помощью IsolationForest:")
df_cleaned_isolation
```
